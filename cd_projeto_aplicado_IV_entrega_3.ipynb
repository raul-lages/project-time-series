# **Referencial Teórico**

Este projeto se posiciona na interseção da econometria e da ciência de dados, comparando modelos de diferentes naturezas para a previsão do desemprego. A seguir, são definidos os conceitos centrais e discutidos os trabalhos correlacionados que fundamentam esta análise.

**Modelos Econométricos Tradicionais**

- ARIMA (Autoregressive Integrated Moving Average): É um dos modelos mais clássicos para previsão de séries temporais, formalizado na metodologia de Box e Jenkins (1970). Ele combina três componentes: um componente autorregressivo (AR), que modela a relação entre uma observação e um número de observações defasadas; um componente de médias móveis (MA), que modela o erro da previsão como uma combinação linear de erros passados; e o componente integrado (I), que utiliza a diferenciação dos dados para tornar a série temporal estacionária. O SARIMA (Seasonal ARIMA) é uma extensão direta deste modelo, que adiciona componentes sazonais (P, D, Q) para capturar padrões que se repetem em intervalos fixos. Esta abordagem é um método univariado robusto e amplamente utilizado para modelar e prever séries temporais econômicas que, como a taxa de desemprego, exibem sazonalidade (DAVIDESCU et al., 2021).

- VAR (Vector Autoregressive): O modelo VAR, introduzido na macroeconomia por Sims (1980), é uma generalização do modelo autorregressivo para múltiplas séries temporais. Em um sistema VAR, cada variável é modelada como uma função linear de seus próprios valores passados e dos valores passados de todas as outras variáveis do sistema. Isso o torna particularmente útil para descrever a dinâmica de interdependência entre variáveis macroeconômicas.

**Modelos de Machine Learning**

- XGBoost (Extreme Gradient Boosting): É um algoritmo de ensemble baseado em árvores de decisão que se destaca pela sua alta performance e eficiência (CHEN; GUESTRIN, 2016). Ele constrói um modelo preditivo de forma sequencial, onde cada nova árvore corrige os erros do conjunto de árvores anterior. Para séries temporais, o XGBoost é utilizado transformando o problema de previsão em um de regressão supervisionada.

- Prophet: Desenvolvido pelo Facebook, o Prophet é um procedimento de previsão baseado em um modelo aditivo decomponível (TAYLOR; LETHAM, 2017), onde tendências não-lineares são ajustadas com sazonalidades e efeitos de feriados. Foi projetado para ser robusto e seus parâmetros são facilmente interpretáveis, permitindo que analistas com conhecimento de domínio ajustem o modelo de forma intuitiva.

- LSTM (Long Short-Term Memory): As redes LSTM são um tipo avançado de rede neural recorrente (RNN) projetadas para aprender dependências de longo prazo em dados sequenciais (HOCHREITER; SCHMIDHUBER, 1997). Por meio de mecanismos de "portões" (gates) que controlam o fluxo de informação, as LSTMs podem reter informações relevantes por longos períodos, tornando-as adequadas para capturar padrões complexos e não-lineares.

**Trabalhos Correlacionados e Alternativas de Solução**

A literatura recente oferece um rico panorama de comparações entre essas abordagens. O trabalho de Athey e Imbens (2019) fornece a justificativa teórica para este projeto, defendendo a integração de métodos de machine learning na pesquisa econômica para alavancar seu poder preditivo. Adicionalmente, revisões de literatura validam a seleção de modelos deste projeto como representativa do "estado da arte" (traduzindo, temos algo como "o que há de mais moderno"), confirmando a relevância da comparação entre abordagens econométricas e de machine learning (SHOBANA; UMAMAHESWARI, 2021).

Estudos empíricos sobre a previsão do desemprego revelam um trade-off claro. Isso evidencia a vantagem potencial de modelos de ML em cenários de quebra estrutural. De forma similar, a pesquisa sobre o mercado de trabalho irlandês, considerando uma amostra de 305 casos (janeiro de 1998 a maio de 2023), destacou o bom desempenho de modelos como XGBoost e Ridge Regression, que possuem mecanismos intrínsecos para evitar sobreajuste (overfitting), uma limitação comum em dados ruidosos (KRISHNAMURTHY, 2023). Na mesma linha, um estudo comparativo sobre a taxa de desemprego no Paquistão validou o uso de regressões regularizadas, como Dynamic Ridge Regression e Lasso Regression, como ferramentas eficazes para modelar o indicador (MUBASHER et al., 2024).

Outros estudos apontam para a importância dos dados de entrada. Especificamente para o Brasil, foco deste trabalho, foi demonstrado que o volume de buscas no Google por termos ligados à procura de emprego (como "seguro desemprego") possui um poder preditivo significativo, sendo capaz de superar a precisão de modelos-base que usam apenas padrões sazonais (LASSO, 2014). Foi demonstrado que a acurácia de modelos como ARIMA e VAR para prever o desemprego em Gana foi significativamente melhorada com a inclusão de dados do Google Trends como variáveis exógenas (ADU et al., 2023). Isso sugere que uma alternativa ou complemento à sofisticação do modelo é a engenharia de atributos com dados não convencionais.

Em domínios análogos, a hierarquia de desempenho se repete. Na previsão de preços de alimentos, modelos de deep learning (como LSTM) superaram o ARIMA, que por sua vez superou o Prophet em acurácia (MENCULINI et al., 2021). No entanto, a vantagem do deep learning veio ao custo de maior complexidade e tempo computacional. Por fim, foi observado na previsão de casos de COVID-19 que a acurácia do Prophet diminuía em horizontes de previsão mais longos em comparação com o ARIMA, destacando que a escolha do melhor modelo pode depender do horizonte temporal da previsão (SATRIO et al., 2021).

**Pré-processamento para Modelagem (Estacionariedade)**

Modelos de séries temporais como o SARIMA exigem que a série seja estacionária (média, variância e autocorrelação constantes ao longo do tempo).

- Testes de Estacionariedade (Série Original): Foram aplicados os testes Dickey-Fuller Aumentado (ADF) e KPSS:

    - ADF: O p-valor foi de 0.54 (muito acima de 0.05). Falhamos em rejeitar a hipótese nula (H0) de que a série não é estacionária.

    - KPSS: O p-valor foi de 0.066 (ligeiramente acima de 0.05), indicando que a série é estacionária.

    - Conclusão: Devido à contradição dos testes e à forte tendência visual, consideramos a série original como não estacionária.

- Transformação (Diferenciação): Para tornar a série estacionária, aplicamos duas diferenciações:

    - Diferenciação Regular (d=1): serie.diff(1) foi aplicada para remover a tendência. O teste ADF (p-valor 0.09) ainda falhou em confirmar a estacionariedade.

    - Diferenciação Sazonal (D=1): serie.diff(12) foi aplicada sobre a série já diferenciada (d=1) para remover a forte sazonalidade de 12 meses.

    - Após ambas as diferenciações (d=1 e D=1), a série resultante (serie_estacionaria) foi aprovada em ambos os testes (ADF p-valor 0.01; KPSS p-valor 0.10), confirmando a estacionariedade. Isso justifica o uso de d=1 e D=1 no modelo SARIMA.

- Análise ACF/PACF: Com a série estacionária, os gráficos de Autocorrelação (ACF) e Autocorrelação Parcial (PACF) foram plotados. Esses gráficos são usados para identificar as ordens dos componentes AR (p, P) e MA (q, Q) do modelo SARIMA. A análise visual dos lags significativos (fora da área azul) nestes gráficos sugere os parâmetros (1,1,1) para a ordem não sazonal e (1,1,1,12) para a ordem sazonal, que serão usados no modelo base.

```python
# Códigos aqui
# Imports para Análise, Pré-processamento e Modelagem
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX # Para o SARIMA manual
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from xgboost import XGBRegressor, plot_importance
import warnings
import glob
import os

# Configurações de visualização
warnings.filterwarnings('ignore')
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Configuração do Pandas para exibir floats
pd.set_option('display.float_format', lambda x: '%.2f' % x)

####################################################################################################################

# 1. Mapeamento de meses para converter 'mar/12' -> '2012-03'
mes_map = {
    'jan': '01', 'fev': '02', 'mar': '03', 'abr': '04', 'mai': '05', 'jun': '06',
    'jul': '07', 'ago': '08', 'set': '09', 'out': '10', 'nov': '11', 'dez': '12'
}

# 2. Definição do caminho e busca dos arquivos
folder_path = os.path.join("data", "New")

## Obs: a ideia é deixar os arquivos csv numa pasta. Caso o exemplo acima não rodar, é só baixar os arquivos no caminho e rodar localmente tipo:
# folder_path = r"C:\Users\Raul\Downloads\ProjetoAplicado\00. Projeto Aplicado 4\00. Base de Dados\New"

file_pattern = os.path.join(folder_path, "*.csv")
csv_files = glob.glob(file_pattern)

# 3. Listas para armazenar os DataFrames processados
taxa_df = None
other_main_dfs = []
timeline_dfs = []

print(f"Iniciando consolidação de {len(csv_files)} arquivos...")

# 4. Loop principal para processar cada arquivo
for file_path in csv_files:
    file_name = os.path.basename(file_path)
    
    try:
        # --- CASO 1: Arquivos 'multiTimeline' (Google Trends) ---
        if file_name.startswith('multiTimeline'):
            df = pd.read_csv(file_path, encoding='utf-8', header=1)
            df['Month'] = pd.to_datetime(df['Month'], format='%Y-%m')
            df = df.set_index('Month')
            df.index.name = 'Date'
            timeline_dfs.append(df)

        # --- CASO 2: Arquivo 'TaxaDesemprego.csv' ---
        elif file_name == 'TaxaDesemprego.csv':
            df = pd.read_csv(file_path, encoding='latin-1', sep=';')
            df = df.dropna(axis=1, how='all') 
            date_col_name = df.columns[0]
            
            # Converte a data de 'mar/12' para '2012-03'
            df[date_col_name] = df[date_col_name].str.strip().str.lower().str.split('/').apply(
                lambda x: f"20{x[1]}-{mes_map[x[0]]}"
            )
            df[date_col_name] = pd.to_datetime(df[date_col_name], format='%Y-%m')
            
            df = df.set_index(date_col_name)
            df.index.name = 'Date'
            taxa_df = df

        # --- CASO 3: Outros arquivos ('STI', 'STP') ---
        else: 
            try:
                df = pd.read_csv(file_path, encoding='latin-1', sep=';', skipfooter=1, engine='python')
            except UnicodeDecodeError:
                df = pd.read_csv(file_path, encoding='utf-8', sep=';', skipfooter=1, engine='python')
            
            df = df.dropna(axis=1, how='all')
            
            if df.empty:
                print(f"--- Aviso: Arquivo {file_name} ficou vazio após processamento. Pulando. ---")
                continue 

            date_col_name = df.columns[0] # Pega o nome da coluna de data
            
            # --- LÓGICA CONDICIONAL DE DATA ---
            
            # 1. Pega o primeiro valor da data para verificar o formato
            primeira_data_str = df[date_col_name].iloc[0].strip().lower()
            primeira_parte = primeira_data_str.split('/')[0]

            # 2. Verifica se a primeira parte é texto (ex: 'mar')
            if primeira_parte in mes_map:
                # Se for 'mar', 'abr', etc., usa o mes_map
                df[date_col_name] = df[date_col_name].str.strip().str.lower().str.split('/').apply(
                    lambda x: f"20{x[1]}-{mes_map[x[0]]}"
                )
                df[date_col_name] = pd.to_datetime(df[date_col_name], format='%Y-%m')
            else:
                # Se for '03', '04', etc., usa a conversão numérica
                df[date_col_name] = pd.to_datetime(df[date_col_name], format='%m/%Y')

            # Define a data como o índice
            df = df.set_index(date_col_name)
            df.index.name = 'Date'
            other_main_dfs.append(df)

    except Exception as e:
        print(f"--- Erro ao processar o arquivo {file_name}: {e} ---")

# 5. Consolidação e Processamento Final
print("\nConsolidando DataFrames...")

try:
    df_others = pd.concat(other_main_dfs, axis=1) if other_main_dfs else pd.DataFrame()
    df_timeline = pd.concat(timeline_dfs, axis=1) if timeline_dfs else pd.DataFrame()
    
    if taxa_df is None:
        print("Aviso: 'TaxaDesemprego.csv' não foi encontrado. Prosseguindo sem ele.")
        taxa_df = pd.DataFrame()

    final_dfs_to_concat = []
    if not taxa_df.empty:
        final_dfs_to_concat.append(taxa_df)
    if not df_others.empty:
        final_dfs_to_concat.append(df_others)
    if not df_timeline.empty:
        final_dfs_to_concat.append(df_timeline)

    if not final_dfs_to_concat:
        print("\n--- Erro: Nenhum dado foi lido com sucesso. DataFrame final está vazio. ---")
    else:
        # --- ETAPA DE CONSOLIDAÇÃO ---
        df_final = pd.concat(final_dfs_to_concat, axis=1)
        print("\n--- Consolidação Inicial Concluída ---")
        
        # --- ETAPA DE RENOMEAÇÃO ---
        print("\n--- 1. Renomeando Colunas (Padronização) ---")
        
        column_map = {
            'Taxa de desemprego (desocupação)': 'tx_desemprego_pnad',
            '3695 - Taxa de câmbio - Livre - Dólar americano (compra) - Fim de período - mensal': 'cambio_compra_fim',
            '3696 - Taxa de câmbio - Livre - Dólar americano (venda) - Fim de período - mensal': 'cambio_venda_fim',
            '3697 - Taxa de câmbio - Livre - Dólar americano (compra) - Média de período - mensal': 'cambio_compra_media',
            '3698 - Taxa de câmbio - Livre - Dólar americano (venda) - Média de período - mensal': 'cambio_venda_media',
            '433 - Índice nacional de preços ao consumidor-amplo (IPCA)': 'ipca',
            '4189 - Taxa de juros - Selic acumulada no mês anualizada base 252': 'tx_selic_anualizada',
            '4390 - Taxa de juros - Selic acumulada no mês': 'tx_selic_mensal',
            '29037 - Endividamento das famílias com o Sistema Financeiro Nacional em relação à renda acumulada dos últimos doze meses (RNDBF) - %': 'endividamento_familias',
            '20714 - Taxa média de juros das operações de crédito - Total - % a.a.': 'tx_media_juros_credito_total',
            '20715 - Taxa média de juros das operações de crédito - Pessoas jurídicas - Total - % a.a.': 'tx_media_juros_credito_pj',
            '20716 - Taxa média de juros das operações de crédito - Pessoas físicas - Total - % a.a.': 'tx_media_juros_credito_pf',
            '4466 - Índice nacional de preços ao consumidor-Amplo (IPCA) - Núcleo médias aparadas com suavização - Var. % mensal': 'ipca_nucleo',
            '24363 - Índice de Atividade Econômica do Banco Central - IBC-Br - Índice': 'ibc_br',
            '24364 - Índice de Atividade Econômica do Banco Central (IBC-Br) - com ajuste sazonal - Índice': 'ibc_br_dessazonalizado',
            '29039 - Índice de Commodities - Brasil - Energia (em US Dólares) - Índice': 'ic_energia',
            '29040 - Índice de Commodities - Brasil - Metal (em US Dólares) - Índice': 'ic_metal',
            '29041 - Índice de Commodities - Brasil - Agropecuária (em US Dólares) - Índice': 'ic_agro',
            '29042 - Índice de Commodities - Brasil (em US Dólares) - Índice': 'ic_br',
            '29601 - Índice de Atividade Econômica do Banco Central (IBC-Br) Agropecuária - Índice': 'ibc_br_agro',
            'Seguro desemprego: (Brazil)': 'google_seguro_desemprego',
            'Desemprego + benefício social: (Brazil)': 'google_desempreg_beneficio_social',
            'Vagas de emprego: (Brazil)': 'google_vagas_emprego',
            'Décimo terceiro salário: (Brazil)': 'google_decimo_terceiro_salario',
            'Décimo terceiro: (Brazil)': 'google_decimo_terceiro',
            'Empregos: (Brazil)': 'google_empregos',
            'Vagas: (Brazil)': 'google_vagas',
            'FGTS: (Brazil)': 'google_fgts',
            'Saque FGTS: (Brazil)': 'google_saque_fgts',
            'INSS: (Brazil)': 'google_inss',
            'Aposentadoria: (Brazil)': 'google_aposentadoria'
        }
        
        df_final.rename(columns=column_map, inplace=True)
        
        # --- ETAPA DE LIMPEZA ---
        print("\n--- 2. Correção Crítica dos Tipos de Dados (Vírgula -> Ponto) ---")
        
        cols_to_clean = [
            'endividamento_familias', 
            'tx_media_juros_credito_total',
            'tx_media_juros_credito_pj',
            'tx_media_juros_credito_pf',
            'ipca_nucleo', 'ibc_br', 'ibc_br_dessazonalizado', 
            'ic_energia', 'ic_metal', 'ic_agro', 'ic_br', 'ibc_br_agro'
        ]
        
        # Garantir que apenas colunas existentes sejam processadas
        cols_existentes = [col for col in cols_to_clean if col in df_final.columns]
        
        print(f"Colunas a serem limpas: {cols_existentes}")
        
        for col in cols_existentes:
            # Substitui vírgula por ponto
            df_final[col] = df_final[col].astype(str).str.replace(',', '.', regex=False)
            # Converte para numérico, tratando erros
            df_final[col] = pd.to_numeric(df_final[col], errors='coerce')
        
        # --- VERIFICAÇÃO FINAL ---
        print("\n--- 3. Verificação Final da Estrutura ---")
        print("Informações (df_final.info()):")
        df_final.info()
        
        print("\nEstatísticas Descritivas (df_final.describe()):")
        print(df_final.describe())

        print("\n--- Processamento Completo ---")

except Exception as e:
    print(f"\n--- Erro na consolidação ou processamento final: {e} ---")

# --- Configurações de Exibição do Pandas ---
# Força o pandas a mostrar todas as colunas (sem '...')
pd.set_option('display.max_columns', None)
# Aumenta a largura máxima de exibição para evitar quebra de linha
pd.set_option('display.width', 2000) 

print("\n--- 5 Primeiras Linhas do DataFrame Consolidado (Todas as Colunas) ---")

# Verifica se o df_final existe antes de tentar imprimi-lo
if 'df_final' in locals():
    print(df_final.head())
else:
    print("Erro: O DataFrame 'df_final' não foi encontrado.")
    print("Por favor, execute o chunk de consolidação anterior primeiro.")

# Opcional se for necessário fazer um reset depois:
# pd.reset_option('display.max_columns')
# pd.reset_option('display.width')

####################################################################################################################

# 3.1.1. Série Temporal Completa

import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de plotagem (geralmente feitas no início do notebook)
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (14, 7)

# Plotar a série temporal da variável alvo
sns.lineplot(data=df_final, x=df_final.index, y='tx_desemprego_pnad')

# Adicionar títulos e rótulos
plt.title('Série Temporal da Taxa de Desemprego (PNAD)')
plt.xlabel('Data')
plt.ylabel('Taxa (%)')

# Exibir o gráfico
plt.show()

####################################################################################################################

# 3.1.2. Decomposição (Tendência, Sazonalidade, Resíduo)

from statsmodels.tsa.seasonal import seasonal_decompose

# Definir o período de sazonalidade (12 para dados mensais)
# Seus dados são mensais, então o ciclo sazonal é de 12 meses.
periodo_sazonal = 12

# Realizar a decomposição
# Usamos 'multiplicative' porque flutuações econômicas (como o desemprego)
# muitas vezes têm uma sazonalidade que cresce/diminui com a tendência.
# Se o 'seasonal' plot parecer estranho, tente 'additive'.
decomposicao = seasonal_decompose(df_final['tx_desemprego_pnad'].dropna(), 
                                  model='multiplicative', 
                                  period=periodo_sazonal)

# Plotar os componentes
fig = decomposicao.plot()
fig.set_size_inches(14, 10)
plt.suptitle('Decomposição da Série Temporal (Multiplicativa)', y=1.02)
plt.tight_layout()
plt.show()

####################################################################################################################

# 3.2.1. Testes (ADF, KPSS)

from statsmodels.tsa.stattools import adfuller, kpss

# Criar uma função para executar os testes e imprimir os resultados
def testes_estacionariedade(serie, nome_serie='Série'):
    """
    Executa os testes ADF e KPSS e imprime os resultados de forma clara.
    """
    print(f'--- Testando Estacionariedade para: {nome_serie} ---')
    
    # Teste ADF
    # H0: A série NÃO é estacionária (possui raiz unitária).
    result_adf = adfuller(serie.dropna())
    print('\n--- Teste ADF (Augmented Dickey-Fuller) ---')
    print(f'Estatística ADF: {result_adf[0]:.4f}')
    print(f'p-valor: {result_adf[1]:.4f}')
    if result_adf[1] <= 0.05:
        print("Resultado: Rejeita H0. A série é ESTACIONÁRIA.")
    else:
        print("Resultado: Falha em rejeitar H0. A série NÃO é ESTACIONÁRIA.")

    # Teste KPSS
    # H0: A série É estacionária em torno de uma constante.
    # Usamos 'c' para 'constant' (estacionária em torno da média)
    print('\n--- Teste KPSS ---')
    result_kpss = kpss(serie.dropna(), regression='c', nlags='auto')
    print(f'Estatística KPSS: {result_kpss[0]:.4f}')
    print(f'p-valor: {result_kpss[1]:.4f}')
    # Nota: A interpretação do p-valor do KPSS é invertida
    if result_kpss[1] <= 0.05:
        print("Resultado: Rejeita H0. A série NÃO é ESTACIONÁRIA.")
    else:
        print("Resultado: Falha em rejeitar H0. A série é ESTACIONÁRIA.")
    print('--------------------------------------------------')

# Executar os testes na série original
testes_estacionariedade(df_final['tx_desemprego_pnad'], nome_serie='Original')

####################################################################################################################

# 3.2.2. Análise de Diferenciação

# É muito provável que a série original não seja estacionária.
# Vamos aplicar a primeira diferenciação (d=1) para remover a tendência.
# d=1 calcula a diferença: y_t - y_{t-1}
serie_diff_1 = df_final['tx_desemprego_pnad'].diff(1).dropna()

# Plotar a série com 1ª diferenciação
serie_diff_1.plot(title='Série com 1ª Diferenciação (d=1)')
plt.show()

# Re-testar a estacionariedade
testes_estacionariedade(serie_diff_1, nome_serie='Com 1ª Diferenciação (d=1)')

# --- Análise e Próximo Passo ---
# Se a série 'serie_diff_1' AINDA não for estacionária (comum se houver forte sazonalidade),
# aplicamos a diferenciação sazonal (o 'D' no SARIMA).
# D=1 calcula a diferença: y_t - y_{t-12}

print("\n... Verificando diferenciação sazonal ...")

# Aplicar diferenciação sazonal (D=1) na série JÁ diferenciada (d=1)
serie_diff_1_12 = serie_diff_1.diff(12).dropna()

# Plotar a série com ambas as diferenciações
serie_diff_1_12.plot(title='Série com 1ª Diferenciação (d=1) e Sazonal (D=1)')
plt.show()

# Testar a série final
testes_estacionariedade(serie_diff_1_12, nome_serie='Com d=1 e D=1')

# Guardar a série estacionária para a próxima etapa
# (Assumindo que d=1 e D=1 foram necessários)
serie_estacionaria = serie_diff_1_12

####################################################################################################################

# 3.3. Análise de Autocorrelação (ACF/PACF)
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Certifique-se de usar a série que se tornou estacionária nos testes anteriores
# (Ex: serie_estacionaria = serie_diff_1_12)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

# Gráfico ACF (Autocorrelation Function)
# Ajuda a determinar a ordem MA (q) e MA Sazonal (Q)
plot_acf(serie_estacionaria, lags=40, ax=ax1)
ax1.set_title('Função de Autocorrelação (ACF)')

# Gráfico PACF (Partial Autocorrelation Function)
# Ajuda a determinar a ordem AR (p) e AR Sazonal (P)
plot_pacf(serie_estacionaria, lags=40, ax=ax2)
ax2.set_title('Função de Autocorrelação Parcial (PACF)')

plt.tight_layout()
plt.show()

####################################################################################################################

# --- 3.4.1. Matriz de Correlação (Heatmap e Análise Quantitativa) ---
print("--- Iniciando Análise de Correlação ---")

# Calcular a matriz de correlação
# Usamos o df_final original, pois queremos a correlação das variáveis
corr_matrix = df_final.corr()

# --- Parte 1: Visual (Heatmap) ---
print("Gerando Heatmap...")
plt.figure(figsize=(25, 20)) # Aumentar o tamanho para acomodar ~30 features

# Plotar o heatmap
sns.heatmap(corr_matrix, 
            annot=True,       # Mostrar os valores de correlação
            cmap='coolwarm',  # Esquema de cores (divergente)
            fmt='.2f',        # Formatar os números para 2 casas decimais
            linewidths=0.5,
            annot_kws={"size": 7}) # Diminuir o tamanho da fonte da anotação

plt.title('Matriz de Correlação de Todas as Variáveis', fontsize=20)
plt.show()


# --- Parte 2: Textual (Valores Ordenados) ---

print("\n" + "="*60)
print(" ANÁLISE DE CORRELAÇÃO (Valores Ordenados)")
print("="*60 + "\n")

# --- 2.1. Correlação com o Alvo (tx_desemprego_pnad) ---
target_var = 'tx_desemprego_pnad'

if target_var in corr_matrix:
    print(f"--- 1. Maiores Correlações com a Variável Alvo ({target_var}) ---")
    
    # Pega a coluna da variável alvo e ordena pelo valor absoluto (para ver os mais fortes)
    # mas mantém o valor original (para ver se é positivo ou negativo)
    target_corr = corr_matrix[target_var].sort_values(key=lambda x: x.abs(), ascending=False)
    
    # Remove a própria variável (correlação 1.0)
    target_corr = target_corr.drop(target_var, errors='ignore') 
    
    print(target_corr)
    print("\n")
    
else:
    print(f"Aviso: Variável alvo '{target_var}' não encontrada na matriz de correlação.")
    print("\n")


# --- 2.2. Multicolinearidade (Maiores Correlações ENTRE Features) ---
print("--- 2. Maiores Correlações (Multicolinearidade) entre Features ---")
print("(Mostrando os 20 pares mais fortes, ignorando auto-correlação de 1.0)")
print("(Nota: Pares (A,B) e (B,A) são mostrados, por isso 20 linhas = 10 pares)\n")

# Pega o valor absoluto para ordenar pela força da correlação
corr_abs = corr_matrix.abs()

# "Desempilha" (unstack) a matriz para uma lista de pares
stacked_corr = corr_abs.stack()

# Ordena os pares do maior para o menor
sorted_corr = stacked_corr.sort_values(ascending=False)

# Remove as auto-correlações (ex: ('ipca', 'ipca') == 1.0)
# Compara o nome da feature no nível 0 do índice com o nível 1
unique_pairs = sorted_corr[sorted_corr.index.get_level_values(0) != sorted_corr.index.get_level_values(1)]

# Mostra os 20 primeiros (que serão os 10 pares únicos mais fortes)
print(unique_pairs.head(20))

print("\n" + "="*60)
print(" FIM DA ANÁLISE DE CORRELAÇÃO")
print("="*60 + "\n")

####################################################################################################################

# 3.4.2. Análise de Features Exógenas vs. Alvo (Dual-axis plots)

# Escolhas:
# 1. 'ibc_br_dessazonalizado': Correlação negativa mais forte (-0.81)
# 2. 'tx_selic_anualizada': Correlação negativa forte (-0.58) e economicamente muito relevante
# 3. 'google_desempreg_beneficio_social': Correlação positiva mais forte (0.37)
# 4; 'ipca_nucleo': correlação negativa de (-0.25)

features_para_analisar = [
    'ibc_br_dessazonalizado', 
    'tx_selic_anualizada', 
    'google_desempreg_beneficio_social',
    'ipca_nucleo'
]
target = 'tx_desemprego_pnad'

# O loop de plotagem
for feature in features_para_analisar:
    
    # Criar a figura e o primeiro eixo (ax1)
    fig, ax1 = plt.subplots(figsize=(14, 7))
    
    # Plotar o Desemprego (eixo da esquerda)
    ax1.plot(df_final.index, df_final[target], color='blue', label='Desemprego')
    ax1.set_xlabel('Data')
    ax1.set_ylabel(target, color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    # Criar o segundo eixo (ax2) que compartilha o eixo X
    ax2 = ax1.twinx() 
    
    # Plotar a Feature Exógena (eixo da direita)
    ax2.plot(df_final.index, df_final[feature], color='red', label=feature)
    ax2.set_ylabel(feature, color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    # Adicionar título e legendas
    fig.suptitle(f'{target.upper()} vs. {feature.upper()}', fontsize=16)
    # Combinar legendas dos dois eixos
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc='upper left')
    
    plt.show()
```

# **Modelo base**

Para estabelecer uma baseline de desempenho, foram aplicados dois modelos distintos: um modelo estatístico clássico (SARIMA), focado na própria série temporal, e um modelo de machine learning (XGBoost), que utiliza um conjunto rico de atributos. Até a entrega final, mais modelos serão adicionados.

**Estratégia de Divisão dos Dados**

- Engenharia de Atributos (Feature Engineering): Para o modelo de machine learning, foi necessário transformar a série temporal em um problema de regressão supervisionada. Foi criado um conjunto robusto de 70 atributos, incluindo:

    - Features de Target: Foram criadas features de lags para capturar momentum (lag_1) e sazonalidade (lag_12), além de janelas móveis para tendência de curto prazo (roll_mean_3) e volatilidade anual (roll_std_12).

    - Features Exógenas: O conjunto de 29 features exógenas foi curado para um conjunto robusto e de baixa redundância de 5 variáveis com alta correlação com o alvo (ex: ibc_br, tx_selic, google_seguro_desemprego), todas defasadas em 1 período (lag_1) para evitar vazamento de dados (data leakage).

    - Features de Data (Sazonais): A feature month (mês) foi transformada em duas componentes cíclicas (month_sin e month_cos). Isso ajudou o modelo a entender a sazonalidade de forma contínua (onde dezembro está "próximo" de janeiro) e melhorou a performance que a feature categórica month estava prejudicando.

- Divisão Temporal (Train/Validation/Test): A série foi dividida em três conjuntos distintos, mantendo a ordem cronológica:

    - Treino: 126 observações (Mar/2013 a Ago/2023). Este conjunto foi usado para o treinamento inicial dos modelos e para a busca de hiperparâmetros no GridSearchCV.

    - Validação: 12 observações (Set/2023 a Ago/2024). Este conjunto foi usado para avaliar o overfitting e selecionar os melhores hiperparâmetros.

    - Teste: 12 observações (Set/2024 a Ago/2025). Este conjunto foi mantido "cego" (nunca visto pelo modelo) para a avaliação final de performance e generalização.

- Escalonamento (Scaling): Foi utilizado o StandardScaler para normalizar as features numéricas.
    - Metodologia: O scaler foi ajustado (fit) apenas nos dados de Treino (126 observações). Em seguida, ele foi usado para transformar os conjuntos de Treino, Validação e Teste. Isso garantiu que a média e o desvio padrão dos períodos futuros não "vazassem" para o processo de treinamento.

**Modelo Base 1: SARIMA (Modelo Estatístico)**

- Método Aplicado: O SARIMA é um modelo estatístico ideal para séries que apresentam tendência e sazonalidade claras, como a do desemprego. Com base na análise de estacionariedade (que sugeriu d=1, D=1) e nos gráficos ACF/PACF, foi definido um modelo base com os parâmetros order=(1, 1, 1) (não sazonal) e seasonal_order=(1, 1, 1, 12) (sazonal), com período de 12 meses.

- Análise dos Resultados: O modelo foi treinado nos 126 pontos de treino e usado para prever os 12 meses de validação. Posteriormente, foi re-treinado com os dados de treino+validação (138 pontos) para fazer a previsão final no conjunto de teste.

    - Validação: O modelo alcançou um MAPE de 6.17% e um RMSE de 0.50.

    - Teste: No conjunto de teste, o desempenho se manteve estável, com um MAPE de 9.44% e um RMSE de 0.65. Os gráficos mostram que o SARIMA capturou a tendência de queda corretamente, embora com uma ligeira superestimação dos valores.

O SARIMA provou ser um baseline robusto. A diferença entre o erro de validação (0.50) e o de teste (0.65) é pequena, indicando que o modelo generaliza bem e não sofre de overfitting. Seus diagnósticos de resíduos indicaram que o modelo capturou bem a auto-correlação e a sazonalidade. Qualquer modelo de machine learning teria que superar significativamente o RMSE de 0.65 para justificar sua complexidade.

**Modelo Base 2: XGBoost (Modelo de Machine Learning)**

- Método Aplicado: Foi utilizado um modelo eXtreme Gradient Boosting (XGBoost), uma abordagem multivariada que utiliza todas as features criadas (lags, janelas móveis, exógenas e de data).

- Otimização: Foi usado o GridSearchCV para encontrar os melhores hiperparâmetros, com foco em combater o overfitting (ex: max_depth: [3, 4, 5], reg_lambda: [1, 5, 10]).

- Validação Cruzada: A validação cruzada dentro do GridSearchCV foi feita com o TimeSeriesSplit (com 3 divisões), que respeita a ordem cronológica dos dados, treinando em dados mais antigos e validando em dados mais recentes.

- Metodologia Final: Após o GridSearchCV identificar os melhores parâmetros usando o conjunto de Treino, um modelo final foi instanciado com esses parâmetros e re-treinado nos dados de Treino + Validação. A previsão no conjunto de Teste foi feita usando este modelo final, garantindo uma comparação direta com a metodologia do SARIMA.

- Análise dos Resultados:

    - Validação: O XGBoost apresentou um desempenho excelente no conjunto de validação, com um MAPE de apenas 1.71% e um RMSE de 0.15.

    - Teste: No entanto, ao ser avaliado no conjunto de teste, o desempenho caiu drasticamente. O MAPE subiu para 10.49% e o RMSE para 0.77.

    - Análise de Overfitting: A comparação direta entre o desempenho de validação (MAPE 1.71%) e teste (MAPE 10.49%) indica que o modelo, em sua configuração base, sofreu de forte overfitting. Ele se especializou em prever os dados de validação, mas não conseguiu generalizar seu aprendizado para o conjunto de teste. O gráfico de previsão ilustra isso: o modelo acerta quase perfeitamente a validação, mas erra significativamente no teste.

```python
# 4.1. Feature Engineering (Engenharia de Atributos)

print("Iniciando 4.1 (Revisado para Overfitting)...")

# --- 0. Configuração Inicial ---
df_features = df_final.copy()
target = 'tx_desemprego_pnad'

# --- 1. Features de Data ---
# df_features['month'] = df_features.index.month
df_features['quarter'] = df_features.index.quarter
df_features['year'] = df_features.index.year

# Features Sazonais Cíclicas
month_in_year = df_features.index.month
df_features['month_sin'] = np.sin(2 * np.pi * month_in_year / 12)
df_features['month_cos'] = np.cos(2 * np.pi * month_in_year / 12)

# --- 2. Features de Lag (Alvo) ---
df_features[f'{target}_lag_1'] = df_features[target].shift(1)
df_features[f'{target}_lag_12'] = df_features[target].shift(12)

# --- 3. Features de Janela Móvel (Alvo) ---
df_features[f'{target}_roll_mean_3'] = df_features[target].rolling(window=3).mean().shift(1)
df_features[f'{target}_roll_std_12'] = df_features[target].rolling(window=12).std().shift(1)

# --- 4. Features de Lag (Exógenas) ---
selected_exog_features_CURATED = [
    'ibc_br_dessazonalizado', 
    'ic_agro',
    'tx_selic_anualizada',
    # 'tx_media_juros_credito_pj',
    'google_seguro_desemprego',
    'google_vagas_emprego'
]

print(f"\nCriando 'lag_1' para {len(selected_exog_features_CURATED)} features exógenas (curadas para evitar overfitting)...")

for col in selected_exog_features_CURATED:
    if col in df_final.columns:
        df_features[f'{col}_lag_1'] = df_features[col].shift(1)
    else:
        print(f"Aviso: Coluna '{col}' não encontrada em df_final.")
        
# --- 5. Seleção Final de Features e Limpeza ---
date_cols = date_cols = ['quarter', 'year', 'month_sin', 'month_cos']
target_derived_cols = [
    f'{target}_lag_1',
    f'{target}_lag_12',
    f'{target}_roll_mean_3',
    f'{target}_roll_std_12'
]
exog_lagged_cols = [f'{col}_lag_1' for col in selected_exog_features_CURATED]
final_model_cols = [target] + date_cols + target_derived_cols + exog_lagged_cols
df_model_data_filtered = df_features[final_model_cols]

print(f"\nFormato original do df_final: {df_final.shape}")
print(f"Formato antes do filtro e dropna: {df_features.shape}")
print(f"Total de colunas selecionadas para o modelo (incluindo alvo): {len(final_model_cols)}")

df_model_data = df_model_data_filtered.dropna()

print(f"Formato após feature engineering e dropna: {df_model_data.shape}")
print(f"\n4.1 (Revisado) concluído. Total de features para o modelo: {df_model_data.shape[1] - 1}")

####################################################################################################################

# 4.2. Definição da Estratégia de Validação (Train-Validation-Test Split)

test_size = 12
val_size = 12

train_val_cutoff = len(df_model_data) - test_size - val_size
val_test_cutoff = len(df_model_data) - test_size

train_df = df_model_data.iloc[:train_val_cutoff]
val_df = df_model_data.iloc[train_val_cutoff:val_test_cutoff]
test_df = df_model_data.iloc[val_test_cutoff:]

target = 'tx_desemprego_pnad'
features = [col for col in df_model_data.columns if col != target]

X_train = train_df[features]
y_train = train_df[target]
X_val = val_df[features]
y_val = val_df[target]
X_test = test_df[features]
y_test = test_df[target]

print("--- Formato (Shape) dos Conjuntos ---")
print(f"X_train: {X_train.shape} | y_train: {y_train.shape}")
print(f"X_val:   {X_val.shape} | y_val:   {y_val.shape}")
print(f"X_test:  {X_test.shape} | y_test:  {y_test.shape}")

print("\n--- Período de Datas dos Conjuntos ---")
print(f"Treino: {X_train.index.min().date()} até {X_train.index.max().date()}")
print(f"Val:    {X_val.index.min().date()} até {X_val.index.max().date()}")
print(f"Teste:  {X_test.index.min().date()} até {X_test.index.max().date()}")

####################################################################################################################

# 4.3. Pipeline de Scaling
features_to_scale = [col for col in X_train.columns if col not in ['month', 'quarter']]
scaler = StandardScaler()

print(f"Ajustando (fit) o scaler em {len(features_to_scale)} colunas de X_train...")
scaler.fit(X_train[features_to_scale])

X_train_scaled = X_train.copy()
X_val_scaled = X_val.copy()
X_test_scaled = X_test.copy()

X_train_scaled[features_to_scale] = scaler.transform(X_train[features_to_scale])
X_val_scaled[features_to_scale] = scaler.transform(X_val[features_to_scale])
X_test_scaled[features_to_scale] = scaler.transform(X_test[features_to_scale])

print("\nScaling concluído.")

####################################################################################################################

# 5.1. Métricas e Funções Auxiliares de Avaliação

def calculate_metrics(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred) * 100
    
    return {
        'RMSE': rmse,
        'MAE': mae,
        'MAPE (%)': mape
    }

def plot_forecast(y_true, y_pred, model_name, set_name):
    if not isinstance(y_pred, pd.Series):
        y_pred = pd.Series(y_pred, index=y_true.index)
    
    plt.figure(figsize=(14, 7))
    sns.lineplot(data=y_true, label='Valores Reais', color='blue', marker='o')
    sns.lineplot(data=y_pred, label=f'Previsão {model_name}', color='red', linestyle='--')
    
    plt.title(f'Previsão vs. Real - Modelo: {model_name} (Conjunto de {set_name})')
    plt.xlabel('Data')
    plt.ylabel('Taxa de Desemprego (%)')
    plt.legend()
    plt.show()

####################################################################################################################

# 5.2. Coleta de Resultados

columns = [
    'RMSE_val', 'MAE_val', 'MAPE_val (%)',
    'RMSE_test', 'MAE_test', 'MAPE_test (%)'
]
results_df = pd.DataFrame(columns=columns)
results_df.index.name = 'Modelo'

def add_results(results_df, model_name, metrics_val, metrics_test):
    new_row = {
        'RMSE_val': metrics_val['RMSE'],
        'MAE_val': metrics_val['MAE'],
        'MAPE_val (%)': metrics_val['MAPE (%)'],
        'RMSE_test': metrics_test['RMSE'],
        'MAE_test': metrics_test['MAE'],
        'MAPE_test (%)': metrics_test['MAPE (%)']
    }
    results_df.loc[model_name] = new_row
    return results_df

print("Framework de Coleta de Resultados Inicializado.")

####################################################################################################################

# 6.1. Implementação e Treinamento (SARIMA)

target = 'tx_desemprego_pnad'
train_data_sarima = train_df[target]

order = (1, 1, 1)
seasonal_order = (1, 1, 1, 12) 

print("Iniciando treinamento do modelo SARIMA (para validação)...")
model_sarima_val = SARIMAX(train_data_sarima,
                            order=order,
                            seasonal_order=seasonal_order,
                            enforce_stationarity=False,
                            enforce_invertibility=False)
results_sarima_val = model_sarima_val.fit(disp=False) 
print("\n--- Sumário do Modelo SARIMA (Treino) ---")
print(results_sarima_val.summary())

####################################################################################################################

# 6.2. Avaliação no set de Validação/Teste (SARIMA)

# --- 1. AVALIAÇÃO NO CONJUNTO DE VALIDAÇÃO ---
print("\n--- Avaliação (Validação) SARIMA ---")
n_steps_val = len(y_val)
pred_val_sarima = results_sarima_val.forecast(steps=n_steps_val)
metrics_val_sarima = calculate_metrics(y_val, pred_val_sarima)
print(f"Métricas de Validação (SARIMA): {metrics_val_sarima}")
plot_forecast(y_val, pred_val_sarima, "SARIMA", "Validação")

# --- 2. AVALIAÇÃO NO CONJUNTO DE TESTE ---
print("\n--- Avaliação (Teste) SARIMA ---")
print("Re-treinando modelo SARIMA com dados de treino + validação...")
full_train_data_sarima = pd.concat([train_df[target], val_df[target]])

model_sarima_test = SARIMAX(full_train_data_sarima,
                            order=order,
                            seasonal_order=seasonal_order,
                            enforce_stationarity=False,
                            enforce_invertibility=False)
results_sarima_test = model_sarima_test.fit(disp=False)

n_steps_test = len(y_test)
pred_test_sarima = results_sarima_test.forecast(steps=n_steps_test)
metrics_test_sarima = calculate_metrics(y_test, pred_test_sarima)
print(f"Métricas de Teste (SARIMA): {metrics_test_sarima}")
plot_forecast(y_test, pred_test_sarima, "SARIMA", "Teste")

# --- 3. ADICIONAR RESULTADOS AO DATAFRAME ---
print("\n--- Atualizando DataFrame de Resultados ---")
results_df = add_results(results_df, 
                         "SARIMA_baseline", 
                         metrics_val_sarima, 
                         metrics_test_sarima)
print(results_df)

####################################################################################################################

# 6.3. Análise de Resíduos (SARIMA)

print("\n--- Análise de Resíduos (Modelo de Validação SARIMA) ---")
fig = results_sarima_val.plot_diagnostics(figsize=(15, 12))
plt.suptitle("Diagnóstico de Resíduos do SARIMA (Validação)", y=1.02)
plt.tight_layout()
plt.show()

####################################################################################################################

# 7.1. Configuração do GridSearchCV (XGBoost)

tscv = TimeSeriesSplit(n_splits=3)
print("TimeSeriesSplit (validador CV) criado.")

print("Iniciando GridSearchCV para XGBoost...")
print("Isso pode demorar alguns minutos...")

xgb_grid_model = XGBRegressor(
    objective='reg:squarederror',
    random_state=42
)

param_grid = {
    'learning_rate': [0.05, 0.1],          # Taxas de aprendizado um pouco menores
    'max_depth': [3, 4, 5],                # <-- MUITO IMPORTANTE: Forçar árvores mais rasas (menos complexas)
    'n_estimators': [100, 200],
    
    # --- Parâmetros de Regularização Explícita ---
    'reg_lambda': [1, 5, 10],              # <-- AUMENTAR: Penalidade L2 (força pesos menores)
    'reg_alpha': [0, 0.5, 1],              # <-- AUMENTAR: Penalidade L1 (pode zerar features inúteis)
    
    # --- Parâmetros de Aleatoriedade (para combater overfitting) ---
    'subsample': [0.7, 0.8],               # <-- DIMINUIR: Usar menos dados por árvore
    'colsample_bytree': [0.7, 0.8]         # <-- DIMINUIR: Usar menos features por árvore
}

grid_search = GridSearchCV(
    estimator=xgb_grid_model,
    param_grid=param_grid,
    cv=tscv,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
    verbose=2
)

# EXECUTAR A BUSCA
grid_search.fit(X_train_scaled, y_train)

print("\n--- Busca Concluída ---")
print("Melhores parâmetros encontrados pelo GridSearchCV:")
print(grid_search.best_params_)

####################################################################################################################

# 7.2. XGBoost GridSearch: Avaliação de Validação e Treinamento Final

print("\n--- 7.2: Avaliação (Validação) XGBoost_GridSearch ---")

# 1. Pegar o melhor modelo do grid (treinado APENAS em X_train)
best_xgb_val_model = grid_search.best_estimator_

# 2. Avaliar este modelo no conjunto de VALIDAÇÃO (previsão de 1-12 meses)
pred_val_xgb_grid = best_xgb_val_model.predict(X_val_scaled)
metrics_val_xgb_grid = calculate_metrics(y_val, pred_val_xgb_grid)
print(f"Métricas de Validação (XGBoost_GridSearch): {metrics_val_xgb_grid}")

# Plotar o gráfico de validação
plot_forecast(y_val, pred_val_xgb_grid, "XGBoost_GridSearch", "Validação")

# --- Treinamento Final (para Teste) ---
print("\n--- 7.2: Treinamento FINAL (XGBoost_GridSearch) com (Train + Val) ---")

# 3. Pegar os MELHORES PARÂMETROS encontrados
best_params = grid_search.best_params_

# 4. Instanciar um NOVO modelo final com esses parâmetros
final_xgb_model = XGBRegressor(
    objective='reg:squarederror', 
    random_state=42, 
    **best_params  # Desempacota o dicionário de melhores parâmetros
)

# 5. Combinar os dados de Treino + Validação (JÁ ESCALADOS)
X_full_train_scaled = pd.concat([X_train_scaled, X_val_scaled])
y_full_train = pd.concat([y_train, y_val])

print(f"Treinando modelo final em {X_full_train_scaled.shape[0]} amostras...")

# 6. TREINAR o modelo final nos dados completos
final_xgb_model.fit(X_full_train_scaled, y_full_train)

# --- Avaliação de Teste (do modelo final) ---
print("\n--- 7.2: Avaliação (Teste) do Modelo XGBoost_GridSearch FINAL ---")

# 7. Avaliar o modelo final no conjunto de TESTE (previsão de 1-12 meses)
pred_test_xgb_grid = final_xgb_model.predict(X_test_scaled)
metrics_test_xgb_grid = calculate_metrics(y_test, pred_test_xgb_grid)
print(f"Métricas de Teste (XGBoost_GridSearch): {metrics_test_xgb_grid}")

# Plotar o gráfico de teste
plot_forecast(y_test, pred_test_xgb_grid, "XGBoost_GridSearch", "Teste")

# --- 8. ADICIONAR RESULTADOS AO DATAFRAME ---
print("\n--- Atualizando DataFrame de Resultados ---")
results_df = add_results(results_df, 
                         "XGBoost_GridSearch", 
                         metrics_val_xgb_grid,  # Métricas de Val do modelo treinado em (Train)
                         metrics_test_xgb_grid  # Métricas de Teste do modelo treinado em (Train+Val)
                        )

print(results_df.sort_values(by='MAPE_test (%)'))

# --- 9. ANÁLISE DE FEATURE IMPORTANCE (do modelo final) ---
print("\n--- Análise de Feature Importance (XGBoost_GridSearch FINAL) ---")
plt.figure(figsize=(10, 15))
plot_importance(final_xgb_model, max_num_features=20, height=0.8)
plt.title('Importância das Features (F-score) - XGBoost_GridSearch Final')
plt.show()

####################################################################################################################

# Passos 8 ao 11: ARIMA, VAR, Prophet e LSTM (Comentados)

####################################################################################################################

# 12.1. Comparação Numérica (Tabela de Métricas)

print("--- Tabela Comparativa de Métricas (Modelos Baseline) ---")
results_df_sorted = results_df.sort_values(by='RMSE_test', ascending=True)
print(results_df_sorted)

####################################################################################################################

# --- SEÇÃO 12.2 MODIFICADA ---
# (Atualizado para usar a variável correta 'pred_test_xgb_grid')

# 12.2. Comparação Gráfica (Previsões no Conjunto de Teste)

print("\n--- Comparação Gráfica das Previsões (Conjunto de Teste) ---")

# Converter previsões para Séries do Pandas com o índice de datas correto
pred_test_sarima_series = pd.Series(pred_test_sarima, index=y_test.index)
# ATUALIZADO: Usar 'pred_test_xgb_grid' do modelo final
pred_test_xgb_grid_series = pd.Series(pred_test_xgb_grid, index=y_test.index)

plt.figure(figsize=(14, 7))

# 1. Plotar os Valores Reais
sns.lineplot(data=y_test, label='Valores Reais', color='black', marker='o', 
             linewidth=2.5, markersize=8)

# 2. Plotar a Previsão do SARIMA
sns.lineplot(data=pred_test_sarima_series, label='Previsão SARIMA', 
             color='blue', linestyle='--')

# 3. Plotar a Previsão do XGBoost
# ATUALIZADO: Plotar a série correta
sns.lineplot(data=pred_test_xgb_grid_series, label='Previsão XGBoost_GridSearch', 
             color='red', linestyle='--')

plt.title('Comparação de Previsões vs. Real (Conjunto de Teste)')
plt.xlabel('Data')
plt.ylabel('Taxa de Desemprego (%)')
plt.legend()
plt.show()

####################################################################################################################

# 12.3. Comparação Gráfica (Métricas de Erro)

print("\n--- Comparação Gráfica das Métricas (Conjunto de Teste) ---")

results_test_only = results_df[['RMSE_test', 'MAE_test', 'MAPE_test (%)']]
results_test_long = results_test_only.reset_index().melt(
    id_vars='Modelo', 
    var_name='Métrica', 
    value_name='Valor'
)

g = sns.catplot(
    data=results_test_long,
    x='Modelo',
    y='Valor',
    hue='Modelo',
    col='Métrica',
    kind='bar',
    palette='viridis',
    sharey=False,
    legend=False
)

g.fig.suptitle('Comparação de Métricas de Erro (Conjunto de Teste)', y=1.03, fontsize=16)
g.set_axis_labels("Modelo", "Valor do Erro")
g.set_titles("{col_name}")
plt.tight_layout()
plt.show()

####################################################################################################################

# 12.4 Comparação dos Resíduos 

print("\n--- Comparação dos Resíduos (Conjunto de Teste) ---")

# Converter previsões para Séries do Pandas
sarima_preds_series = pd.Series(pred_test_sarima, index=y_test.index, name='SARIMA_Preds')
# ATUALIZADO: Usar 'pred_test_xgb_grid'
xgb_preds_series = pd.Series(pred_test_xgb_grid, index=y_test.index, name='XGB_Grid_Preds')

# Calcular os resíduos
sarima_residuals = y_test - sarima_preds_series
xgb_residuals = y_test - xgb_preds_series # Esta variável agora usa a previsão correta

# --- 2. Plot dos Resíduos ao Longo do Tempo ---
print("Gerando gráfico de resíduos ao longo do tempo...")
plt.figure(figsize=(16, 7))

plt.plot(sarima_residuals.index, sarima_residuals, label='Resíduos SARIMA', color='blue', alpha=0.7)

# Plotar os resíduos corretos e atualizar o label
plt.plot(xgb_residuals.index, xgb_residuals, label='Resíduos XGBoost_GridSearch', color='red', alpha=0.7)

plt.axhline(y=0, color='black', linestyle='--', linewidth=2, label='Previsão Perfeita (Erro = 0)')
plt.title('Resíduos dos Modelos (Conjunto de Teste)', fontsize=16)
plt.xlabel('Data')
plt.ylabel('Erro (Real - Previsto)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 3. Plot da Distribuição dos Resíduos ---
print("Gerando gráfico de distribuição dos resíduos...")
plt.figure(figsize=(14, 6))

# Gráfico 1: SARIMA
plt.subplot(1, 2, 1)
sns.histplot(sarima_residuals, kde=True, color='blue', bins=30)
plt.title('Distribuição dos Resíduos SARIMA')
plt.xlabel('Valor do Erro')
plt.ylabel('Frequência')
plt.axvline(x=0, color='black', linestyle='--', linewidth=2)

# Gráfico 2: XGBoost
plt.subplot(1, 2, 2)
# ATUALIZADO: Plotar os resíduos corretos e atualizar o título
sns.histplot(xgb_residuals, kde=True, color='red', bins=30)
plt.title('Distribuição dos Resíduos XGBoost_GridSearch')
plt.xlabel('Valor do Erro')
plt.ylabel('Frequência')
plt.axvline(x=0, color='black', linestyle='--', linewidth=2)

plt.suptitle('Distribuição dos Erros dos Modelos', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

print("\n--- Script Concluído com Sucesso ---")
```
